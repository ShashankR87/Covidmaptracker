```python
dfff=pd.DataFrame({'count' : csvfile.groupby( [ "State", "Zone"] ).size()}).reset_index()
```

```python
dfff
```

```python
df3=pd.DataFrame()
```

```python
df3['State']=csvfile['State'].unique().tolist()
```

```python
df3['Green']=0
df3['Orange']=0
df3['Red']=0
```

```python
df3['State'].unique().tolist()
```

```python
df3.set_index("State", inplace=True)
```

```python
for index, row in dfff.iterrows():
    if(row['Zone']=="Green Zone"):
        df3.loc[row['State']].Green=row['count']
    if(row['Zone']=="Orange Zone"):
        df3.loc[row['State']].Orange=row['count']
    if(row['Zone']=="Red Zone"):
        df3.loc[row['State']].Red=row['count']  
```

```python
negreen=df3.loc['Arunachal Pradesh'].Green+df3.loc['Assam'].Green+df3.loc['Manipur'].Green+df3.loc['Nagaland'].Green+df3.loc['Sikkim'].Green+df3.loc['Tripura'].Green+df3.loc['Meghalaya'].Green+df3.loc['Mizoram'].Green
neRed=df3.loc['Arunachal Pradesh'].Red+df3.loc['Assam'].Red+df3.loc['Manipur'].Red+df3.loc['Nagaland'].Red+df3.loc['Sikkim'].Red+df3.loc['Tripura'].Red+df3.loc['Meghalaya'].Red+df3.loc['Mizoram'].Red
neOrange=df3.loc['Arunachal Pradesh'].Orange+df3.loc['Assam'].Orange+df3.loc['Manipur'].Orange+df3.loc['Nagaland'].Orange+df3.loc['Sikkim'].Orange+df3.loc['Tripura'].Orange+df3.loc['Meghalaya'].Orange+df3.loc['Mizoram'].Orange
```

```python
jkgreen=df3.loc['Ladakh'].Green+df3.loc['Jammu And Kashmir'].Green
jkOrange=df3.loc['Ladakh'].Orange+df3.loc['Jammu And Kashmir'].Orange
jkRed=df3.loc['Ladakh'].Red+df3.loc['Jammu And Kashmir'].Red
```

```python
df3.drop(['Arunachal Pradesh', 'Assam','Manipur','Nagaland','Sikkim','Tripura','Meghalaya','Mizoram','Chandigarh','Lakshadweep','Daman And Diu','Dadra And Nagar Haveli','Puducherry','Ladakh','Jammu And Kashmir'],inplace=True)
```

```python
ljk = pd.DataFrame({"State":['J and K','NorthEast'],"Green":[jkgreen,negreen], 
                    "Orange":[jkOrange,neOrange],
                    "Red":[jkRed,neRed]})
```

```python
ddd=pd.DataFrame(ljk)
```

```python
df3=df3.reset_index()
```

```python
dddd=df3.append(ddd)
```

```python
dddd=dddd.sort_values(by =['State'])
```
