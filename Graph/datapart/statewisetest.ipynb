```python
#download complete.csv from https://www.kaggle.com/imdevskp/covid19-corona-virus-india-dataset
import pandas as pd
```

```python
col_list=["Date","Name of State / UT","Total Confirmed cases","Death"]
```

```python
data = pd.read_csv("complete.csv",usecols=col_list)
```

```python
df = pd.DataFrame(data)
```

```python
uniquestates=df['Name of State / UT'].unique()
```

```python
for i in uniquestates:
    df2=df.loc[df['Name of State / UT']==i]
    df2.to_csv(i+'.csv', index = False, header=True)
```

```python
kerala1=pd.read_csv("kerala.csv")
keraladf = pd.DataFrame(kerala1)
```

```python
combinedarraydfcolheaders=list()
combinedarraydfcolheaders.append('Date')
combinedarraydfcolheaders1=list()
combinedarraydfcolheaders1.append('Date')
```

```python
combineddf=pd.DataFrame()
combineddf['Date']=kerala1['Date']
i=1
for statename in uniquestates:
    combineddf[statename]=0
```

```python
combinedarray=combineddf.to_numpy()  
for statename in uniquestates:
    combinedarraydfcolheaders.append(statename)
    storedstate = pd.read_csv(statename+".csv",usecols=["Date","Total Confirmed cases"])
    storedstatedf=pd.DataFrame(storedstate)
    arrayy=storedstatedf.to_numpy()
    for date in range(0,arrayy.shape[0]):
        for combdate in range(0,combinedarray.shape[0]):
            if(arrayy[date][0]==combinedarray[combdate][0]):
                combinedarray[combdate][i]=arrayy[date][1]
    i+=1
```

```python
combinedarraydf=pd.DataFrame(combinedarray,columns=combinedarraydfcolheaders)
combinedarraydf.to_csv('casesexp.csv', index = False, header=True)                 
```

```python
combineddf1=pd.DataFrame()
combineddf1['Date']=kerala1['Date']
i=1
for statename in uniquestates:
    combineddf1[statename]=0
combinedarray1=combineddf1.to_numpy()  
for statename in uniquestates:
    combinedarraydfcolheaders1.append(statename)
    storedstate1 = pd.read_csv(statename+".csv",usecols=["Date","Death"])
    storedstatedf1=pd.DataFrame(storedstate1)
    arrayy1=storedstatedf1.to_numpy()
    for date in range(0,arrayy1.shape[0]):
        for combdate in range(0,combinedarray1.shape[0]):
            if(arrayy1[date][0]==combinedarray1[combdate][0]):
                combinedarray1[combdate][i]=arrayy1[date][1]
    i+=1
    
combinedarraydf1=pd.DataFrame(combinedarray1,columns=combinedarraydfcolheaders1)
combinedarraydf1.to_csv('deathexp.csv', index = False, header=True)
```
